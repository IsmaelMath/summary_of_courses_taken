### Plots de Regress√£o

```python
import seaborn as sns
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
tips = sns.load_dataset("tips"); tips.head(5)
```

```python
# hue="sex" √© usado para classificar e nesse caso √© em Male e Female
# markers √© usado para definir o formato do objeto
# scatter_kws={"s": 100} Para definir o tamanho dos objetos no gr√°fico

sns.lmplot(x="total_bill", y="tip", data=tips, hue="sex", markers=["o", "v"], scatter_kws={"s":100})
```

```python
# Podemos segregar os dados com no hue, que nesse caso foi male e female usando o parametro col="sex"
sns.lmplot(x="total_bill", y="tip", data=tips, hue="sex",col="sex")
```

```python
# Usando mais formas de segrega√ß√£o de dados agora utilizando a variavel horario
sns.lmplot(x="total_bill", y="tip", data=tips, hue="sex",col="sex", row="time")
```

```python
sns.lmplot(x="total_bill", y="tip", data=tips, hue="sex",col="day")
```

```python
# Podemos mudar o tamanho das figuras usando o parametro aspect=0.6 e size=8
sns.lmplot(x="total_bill", y="tip", data=tips, hue="sex",col="day", aspect=0.6)
```
