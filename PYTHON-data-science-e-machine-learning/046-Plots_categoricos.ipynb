### Plots Categóricos

Agora vamos discutir como usar seaborn para traçar dados categóricos. Existem alguns tipos de argumentos principais para isso:

- factorplot
- boxplot
- violinplot
- stripplot
- swarmplot
- barplot
- countplot

Vamos passar por exemplos de cada um.

```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# Usando esse comando eu não precisa plt.show()
%matplotlib inline
```

```python
tips = sns.load_dataset("tips"); tips.head(5)
```

```python
# Primeiro lidando o gráfico barplot()
sns.barplot(x="sex", y="total_bill", data=tips)
```

```python
sns.barplot(x="sex", y="total_bill", data=tips, estimator=np.std)
```

```python
sns.countplot(x="sex", data=tips)
```

```python
sns.boxplot(x="day", y="total_bill", data=tips, palette="rainbow", hue="sex")
```

```python
sns.boxplot(data=tips, palette="rainbow", orient="h")
```

```python
# Esse modelo de gráfico ele utiliza kde
# Ele usa o modelo não parametrico

sns.violinplot(x="day", y="tip", data=tips, hue="sex", split=True)
```

```python
# Stripplot
sns.stripplot(x="day", y="total_bill", data=tips, jitter=True, hue="sex")
```

```python
sns.swarmplot(x="day", y="total_bill", data=tips, hue="sex")
```

```python
sns.swarmplot(x="day", y="total_bill", data=tips, color="black")
sns.violinplot(x="day", y="total_bill", data=tips)
```

```python
# Nas novas versões do seaborn não existe mais factorplot 
# Existe apenas catplot que substitui factorplot

sns.catplot(x="sex", y="total_bill", data=tips, kind="bar")
```
