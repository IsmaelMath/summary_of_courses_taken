### Estilo de Cores

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
tips = sns.load_dataset("tips"); tips.head()
```

```python
# Gerando um countplot()
sns.set_style("ticks") # É o estilo do gráfico. Podendo ser os seguintes: "darkgrid", "whitegrid", "dark", "white" e "ticks"
plt.figure(figsize=(6,4)) # Definindo o tamanho do grafico
sns.countplot(x="sex", data=tips) # Definindo variaveis para meu gráfico
```

```python
# Gerando um countplot()
sns.set_style("ticks") # É o estilo do gráfico. Podendo ser os seguintes: "darkgrid", "whitegrid", "dark", "white" e "ticks"
plt.figure(figsize=(8,4)) # Definindo o tamanho do grafico
sns.countplot(x="sex", data=tips) # Definindo variaveis para meu gráfico
sns.despine(right=True) # Com esse método consigo retirar as bordas do meu gráfico podemos usar o Métod left=True e right=True
```

```python
# Gerando um plot de regressão 
plt.figure(figsize=(6,4)) # Definindo o tamanho da figura
sns.lmplot(x="total_bill", y="tip", data=tips, aspect=4) # com o parametro aspect=4 podemos usar para o 
# tamanho do gráfico
```

```python
sns.set_context("poster", font_scale=0.5)
sns.lmplot(x="total_bill", y="tip", data=tips)
```

```python
sns.set_context("poster", font_scale=0.7)
sns.lmplot(x="total_bill", y="tip", data=tips, aspect=4)
```

```python
sns.set_context("poster", font_scale=0.7)
sns.lmplot(x="total_bill", y="tip", data=tips, hue="sex", palette="magma", aspect=4) # O método palette="magma" está relacionado 
# com as cores dos dados no gráfico
```
