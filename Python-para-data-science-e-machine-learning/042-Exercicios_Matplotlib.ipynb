## Data

## Exercício 1

** Acompanhe estes passos: **
* ** Crie um objeto de figura chamado fig usando plt.figure () **
* ** Use add_axes para adicionar um eixo à tela de figura em [0,0,1,1]. Chame esse novo eixo de "ax". **
* ** Plote (x, y) nesses eixos e defina os rótulos e títulos para corresponder ao gráfico abaixo: **

```python
# Criando o objeto fig para usarmos os parametros nos gráficos a seguir
fig = plt.figure(figsize=(4,3), dpi=100)

plt.subplot(1,2,1)
# Aqui definimos o tamanho a fonte da figura como um todo (font.size, xtick.labelsize, ytick.labelsize)
plt.rcParams.update({'font.size': 6})

# Aqui utilizamos para adcionar um eixo

# Aqui estamos definindo o tamanho de um subplot em um mesmo gráfico que iremos usar na questão a seguir
axe = fig.add_axes([0,0,1,1])

# Variaveis, largura_da_linha, cor_da_linha
axe.plot(x,y, linewidth=0.7, color="blue")

# Parte de legenda do gráfico
axe.set_xlabel("X")
axe.set_ylabel("Y")
axe.set_title("Título")


# Mudando a posição dos eixos
axe.set_xlim(0,100)
axe.set_ylim(0,200)
```

## Exercício 2
**Crie um objeto de figura e coloque dois eixos sobre ele, ax1 e ax2. Localizado em [0,0,1,1] e [0,2,0,5, .2, .2], respectivamente.**

```python
# Criando o objeto
fig = plt.figure(figsize=(6,4))

# Passando as posições de cada gráfico
ax = fig.add_axes([0,0,1,1])
ax1 = fig.add_axes([0.2,0.5,.2,.2])

```

**Agora plote (x, y) em ambos os eixos. E chame seu objeto de figura para mostrá-lo.**

```python
# Criando o objeto
fig = plt.figure(figsize=(6,4))

# Passando as posições de cada gráfico
ax = fig.add_axes([0,0,1,1])
ax1 = fig.add_axes([0.2,0.5,.2,.2])

# Declarando as variaveis em seus respectivos plots
ax.plot(x,y, color="brown",linestyle="-")
ax1.plot(y,x, color="brown",linestyle="-")

# Passando a legenda para o gráfico principal
ax.set_xlabel("Eixo x")
ax.set_ylabel("Eixo y") 
ax.set_title("Gráfico com a função de Subplot")

# Definindo o tamanho de cada eixo e nesse caso do ax
ax.set_xlim(0,100)
ax.set_ylim(0,200)

# Definindo o tamanho de cada eixo e nesse caso do ax1
ax1.set_xlim(0,200)
ax1.set_ylim(0,100)
```
