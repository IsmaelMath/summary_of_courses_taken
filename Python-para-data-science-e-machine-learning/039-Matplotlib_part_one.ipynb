### Matplotlib

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Criando um array
x = np.linspace(0, 5, 11); x
```

```python
# Criando uma função para obter os valores de y
y = x * x; y
```

```python
# Forma mais facil de chamar o plot
plt.figure(figsize=(6,3)) # Para definir qual o tamanho da figura
plt.plot(y, "yo--") # Fazendo o plot em relação a seus dados e nesse caso recebe os valores y. Observe que podemos
                    # mudar a forma da linha do gráfico.
plt.xlabel("Valores do eixo x") # Título do eixo x
plt.ylabel("Valores do eixo y") # Título do eixo y
plt.legend("Função Y")
plt.title("Meu primeiro gráfico", loc="center") # Título do grafico
```

```python
plt.subplot(1,2,1) # nrows inplica em número de linhas e ncols em quantidade de colunas. 
                                  # O outro agumento igual 1 significa que apartir desse momento você ira 
                                  # trabalhar com o primeiro grafico.
plt.plot(x,y, "r.") # Argumentos do primeiro gráfico

plt.subplot(1,2,2) # Agora iniciamos a trabalhar com segundo gráfico
plt.plot(y,x, "go-") # Argumentos do segundo gráfico
```

```python
fig = plt.figure(figsize=(6,3)) # Criando isso como uma especie de caixa vazia para receber 
                                # argumentos e multiplos gráficos
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.3, 0.3])# [posição_fig, altura_fig, lagura_graf, comprimento_graf]

axes1.plot(x,y)
axes2.plot(y, x)


axes1.set_xlabel("Eixo x") # Legendas dos eixos
axes1.set_ylabel("Eixo y") 

axes1.set_title("Gráfico com a função de Subplot")
```

```python
fig = plt.figure(figsize=(6,3)) # Criando isso como uma especie de caixa vazia.
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
#axes2 = fig.add_axes([0.2, 0.5, 0.3, 0.3])

axes1.plot(x,y)
axes1.plot(y, x) # Observe o primeiro plot da linha de comando anterior. Nesse caso adicionei outra 
                 # função no mesmo grafico


axes1.set_xlabel("Eixo x")
axes1.set_title("Gráfico com duas funções")
```
