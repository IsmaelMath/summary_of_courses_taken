### Matplotlib Part Two

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Criando um array
x = np.linspace(0, 5, 11); x
```

```python
# Criando uma função para obter os valores de y
y = x * x; y
```

```python
# Outra maneira de fazer plots

fig, ax = plt.subplots()

ax.plot(x, x**3, "bo--")
ax.plot(x, x**4, "ro-.")

ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_title("Título")

```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
```

```python
# Criando um for para percorrer array, essa ideia já temos em mente porque nesse caso estamos lidando com subplots
# e estamos tratando de um array

for axe in ax:
    axe.plot(x, y, "bo--")
    axe.set_title("Título")
fig # Estamos chamando para fazer o plot é como se fosse uma especie equivalente a uma função
```

```python
# Essa forma de plotagem ela pode ser feita pq quando se difine plt.subplots(nrows, ncols) estamos 
# tratando de uma array.

ax[0].plot(x,y, "b-.")
ax[0].set_title("Título")


ax[1].plot(x,y, "b-.")
ax[1].set_title("Título")

fig # Novamente usando o chamado 
```

```python
fig, ax = plt.subplots(nrows=5, ncols=3)
# Observe que os gráficos estão se sobre ponto um em cima do outro e com método que 
# vamos ver a seguir evitaremos isso.
```

```python
fig, ax = plt.subplots(nrows=5, ncols=3)
plt.tight_layout() # Agora os espaçamentos estão mais convidativos 
```

```python
# Aqui definimos o tamanho da figura que será padrão em plots futuros depois dessa linha de comando
# podemos mudar a medida da necessidades futuras.
fig = plt.figure(figsize=(8,4), dpi=100)
```

```python
fig, axes = plt.subplots(figsize=(5,4))


# Uma obsevação interessante é que podemos fazer o deslocamento dos dados no gráfico axes.plot(8-x, y, "bo--")
# será deslocados oito unidades para frente no eixo x è equivalente para y que se movera para cima
# isso sera interesante quando formos tratar de grafico de barras

axes.plot(x,y, "bo--")
axes.plot(y, x, "ro--")

axes.set_title("Título")
```
