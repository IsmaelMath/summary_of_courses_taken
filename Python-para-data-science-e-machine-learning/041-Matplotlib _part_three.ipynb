```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Criando um array

x = np.linspace(1,10,11);x
```

```python
# Criando a função que utilizamos na aula anterior
y = x ** 2; y
```

```python
fig, ax = plt.subplots(figsize=(10,8))

# Podemos pasar o método color para escrever o nome dá cor em questão 
# Podemos também usar o método linewidth para definir a largura da linha e podemos chamar esse método de lw
# Podmeos também definir a transparencia da linha com o parametro alpha
# Quando estivermos lidando com color onde não conseguimos incluir o estilo podemos usar linestyle="--"
ax.plot(x, x**2, color="black", linewidth=4, alpha=0.5, linestyle="--")
ax.plot(x, x**3, "yo--", linewidth=2.5)
ax.set_title("Gráfico com diferentes atributos de modificação", loc="center")
ax.set_xlabel("Valores do eixo X")
ax.set_ylabel("Valores do eixo Y")
ax.legend(["Variável X", "Variável Y"])

ax.set_xlim([0,2])
ax.set_ylim([0,10])


# Para visualizar melhor set_xlim() devemos atribuir mais valores a x
```
